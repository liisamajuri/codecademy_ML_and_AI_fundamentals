# Oulu City Print Volumes

## Data Set Description:
The data set  has been retrieved from the Open Data service **Avoindata.fi** maintained by the Digital and Population Data Services Agency (_Digi- ja väestötietovirasto_), with the City of Oulu as its administrator: [**_"Oulun kaupungin tulostusmäärät vuodesta 2019 lähtien"_**](https://www.avoindata.fi/data/fi/dataset/tulostusmaarat-vuodesta-2019-lahtien).

The data set contains printing volumes of the City of Oulu's departments on a monthly basis from 2019 to 2022. The data set is updated once a year. The original data is in [CSV format](https://data.ouka.fi/data/dataset/3f7ba58e-e42e-4c1c-af01-0bebc71418ac/resource/a22d9a43-f519-4403-a346-e9c15cced13c/download/avoindata_tulosmaarat.csv), and the latest update was on February 10, 2023.

The data set includes the following information:

- `Vuosi`: Year (2019-2022)
- `Kuukausi`: Month (tammikuu-joulukuu)
- `Kuukausi_nro`: Month number (1-12)
- `Toimintayksikkö`: The operational unit / business unit of the City of Oulu
- `Mustavalkoinen_A4`: Number of printed A4 black and white pages
- `Värillinen_A4`: Number of printed A4 color pages
- `Mustavalkoinen_A3`: Number of printed A3 black and white pages
- `Värillinen_A3`: Number of printed A3 color pages
- `Yhteensä`: Total number of printed pages

**Note**: The column names mentioned above differ from the column names in the original CSV file. The column names have been shortened/made more descriptive in the early stages of the data handling to facilitate easier handling in later processing.


## Defining project objectives (selection of analysis targets):
        

**1.) Analysis of total number of prints by year and month:**

    - Yearly trend development/distribution
    - Monthly variations


**2.) Analysis of different types of prints:**

    - Proportions of different print types in the total print volumes for different years
    - Comparison of different paper sizes and investigation of possible association
    - Comparison of different color types and investigation of possible association


**3.) Analysis of print quantities by operational/business units:**

    - Yearly total print quantities for each operational/business unit
    - Monthly comparison of average print quantities for each operational/business unit
    - Comparison of operational/business units with highest and lowest print quantities
    - Differences in operational/business units' print types
    - Differences in operational/business units' statistical key figures

## 1.) Analysis of total number of prints by year and month:

```python
# Creating a Pivot table of total printings of each month per year 

if __name__ == "__main__":
    month_order = ['tammikuu', 'helmikuu', 'maaliskuu', 'huhtikuu', 'toukokuu', 'kesäkuu', 'heinäkuu', 'elokuu', 'syyskuu', 'lokakuu', 'marraskuu', 'joulukuu']
    printing_data['Kuukausi'] = pd.Categorical(printing_data['Kuukausi'], categories=month_order, ordered=True)
    
    unpivoted = printing_data.groupby(["Kuukausi", "Vuosi"])["Yhteensä"].sum().reset_index()
    pivoted = unpivoted.pivot_table(index="Kuukausi", columns="Vuosi", values="Yhteensä", aggfunc='sum', margins=True, margins_name='Yhteensä')
    pivoted.replace(np.nan, 0, inplace=True)
    pivoted = pivoted.astype(int)
    
    display(pivoted)
```

```python
# Analyzing yearly trend development/distribution of total printings

if __name__ == "__main__":
    year_sums = printing_data.groupby('Vuosi')["Yhteensä"].sum().reset_index()
    year_sums.rename(columns={"Yhteensä": "Tulosteet kpl"}, inplace=True)
    total_sum = year_sums["Tulosteet kpl"].sum()
    year_sums["Tulosteet %"] = round(year_sums["Tulosteet kpl"] / total_sum * 100, 1)
    display(year_sums)
```

## 2.) Analysis of different types of prints:

### Proportions of different print types in the total print volumes for different years

```python
# Analysing the distribution of different print types in total number of prints per year

if __name__ == "__main__":
 
    # Total sums of different print types per year
    
    columns = ['Vuosi', 'Mustavalkoinen_A4', 'Värillinen_A4', 'Mustavalkoinen_A3', 'Värillinen_A3']
    print_types = printing_data[columns]

    print_types = print_types.groupby('Vuosi').sum().reset_index()

    display(print_types)
    
    # The proportions of the different print types in the total annual print volumes
    
    print_types_proportions = print_types.copy()
    total_prints = print_types_proportions.sum(axis=1)
    
    for column in columns[1:]:
        print_types_proportions[column] = print_types_proportions[column] / total_prints * 100
        print_types_proportions[column] = print_types_proportions[column].round(1)

    display(print_types_proportions)
    
```

### Comparison of operational/business units with highest and lowest print quantities

```python
# Comparing the operational/business units with highest and lowest print volumes using pivot table

if __name__ == "__main__":
    
    unit_totals = printing_data.groupby("Toimintayksikkö")["Yhteensä"].sum().reset_index()
    high_and_low_yearly = unit_totals[(unit_totals['Yhteensä'] == unit_totals['Yhteensä'].max()) | (unit_totals['Yhteensä'] == unit_totals['Yhteensä'].min())]
    high_and_low_yearly.reset_index()

    lowest_unit = high_and_low_yearly.iloc[0]["Toimintayksikkö"]
    highest_unit =high_and_low_yearly.iloc[-1]["Toimintayksikkö"]

    lowest_unit_data = printing_data[printing_data["Toimintayksikkö"] == lowest_unit]
    highest_unit_data = printing_data[printing_data["Toimintayksikkö"] == highest_unit]

    units = [lowest_unit, highest_unit]

    total_prints = pd.Series([lowest_unit_data['Yhteensä'].sum(), highest_unit_data['Yhteensä'].sum()], index=['Lowest printing unit', 'Highest printing unit'])
    prints_year_mean = pd.Series([lowest_unit_data['Yhteensä'].sum() / lowest_unit_data['Vuosi'].nunique(), highest_unit_data['Yhteensä'].sum() / highest_unit_data['Vuosi'].nunique()], index=['Lowest printing unit', 'Highest printing unit'])
    prints_month_mean = pd.Series([lowest_unit_data['Yhteensä'].mean(), highest_unit_data['Yhteensä'].mean()], index=['Lowest printing unit', 'Highest printing unit'])
    black_white_sum = pd.Series([lowest_unit_data['Mustavalkoinen_A4'].sum() + lowest_unit_data['Mustavalkoinen_A3'].sum(), highest_unit_data['Mustavalkoinen_A4'].sum() + highest_unit_data['Mustavalkoinen_A3'].sum()], index=['Lowest printing unit', 'Highest printing unit'])
    black_white_mean = pd.Series([(lowest_unit_data['Mustavalkoinen_A4'].sum() + lowest_unit_data['Mustavalkoinen_A3'].sum()) / len(lowest_unit_data), (highest_unit_data['Mustavalkoinen_A4'].sum() + highest_unit_data['Mustavalkoinen_A3'].sum()) / len(highest_unit_data)], index=['Lowest printing unit', 'Highest printing unit'])
    colour_sum = pd.Series([lowest_unit_data['Värillinen_A4'].sum() + lowest_unit_data['Värillinen_A3'].sum(), highest_unit_data['Värillinen_A4'].sum() + highest_unit_data['Värillinen_A3'].sum()], index=['Lowest printing unit', 'Highest printing unit'])
    colour_mean = pd.Series([(lowest_unit_data['Värillinen_A4'].sum() + lowest_unit_data['Värillinen_A3'].sum()) / len(lowest_unit_data), (highest_unit_data['Värillinen_A4'].sum() + highest_unit_data['Värillinen_A3'].sum()) / len(highest_unit_data)], index=['Lowest printing unit', 'Highest printing unit'])
    A4_sum = pd.Series([lowest_unit_data['Mustavalkoinen_A4'].sum() + lowest_unit_data['Värillinen_A4'].sum(), highest_unit_data['Mustavalkoinen_A4'].sum() + highest_unit_data['Värillinen_A4'].sum()], index=['Lowest printing unit', 'Highest printing unit'])
    A4_mean = pd.Series([(lowest_unit_data['Mustavalkoinen_A4'].sum() + lowest_unit_data['Värillinen_A4'].sum()) / len(lowest_unit_data), (highest_unit_data['Mustavalkoinen_A4'].sum() + highest_unit_data['Värillinen_A4'].sum()) / len(highest_unit_data)], index=['Lowest printing unit', 'Highest printing unit'])
    A3_sum = pd.Series([lowest_unit_data['Mustavalkoinen_A3'].sum() + lowest_unit_data['Värillinen_A3'].sum(), highest_unit_data['Mustavalkoinen_A3'].sum() + highest_unit_data['Värillinen_A3'].sum()], index=['Lowest printing unit', 'Highest printing unit'])
    A3_mean = pd.Series([(lowest_unit_data['Mustavalkoinen_A3'].sum() + lowest_unit_data['Värillinen_A3'].sum()) / len(lowest_unit_data), (highest_unit_data['Mustavalkoinen_A3'].sum() + highest_unit_data['Värillinen_A3'].sum()) / len(highest_unit_data)], index=['Lowest printing unit', 'Highest printing unit'])

    highest_and_lowest_prints = pd.DataFrame({'operational/business unit': units,
                                              'total prints': total_prints.map(int),
                                              'prints/month mean': prints_month_mean.map('{:.2f}'.format),
                                              'prints/year mean': prints_year_mean.map('{:.2f}'.format),
                                              'A3 mean': A3_mean.map('{:.2f}'.format),
                                              'A3 sum': A3_sum.map(int),
                                              'A4 mean': A4_mean.map('{:.2f}'.format),
                                              'A4 sum': A4_sum.map(int),
                                              'black and white mean': black_white_mean.map('{:.2f}'.format),
                                              'black and white sum': black_white_sum.map(int),
                                              'colour mean': colour_mean.map('{:.2f}'.format),
                                              'colour sum': colour_sum.map(int)})

    highest_and_lowest_prints_pivot = highest_and_lowest_prints.transpose()

    display(highest_and_lowest_prints_pivot)
```

```python
# Investigating further one key figure (mean) and differences in operational/business units' mean values

if __name__ == "__main__":
    
    # Creating a table for comparing min and max MEAN values of print types and total sum
    
    unit_means_columns = ['Toimintayksikkö', 'Mustavalkoinen_A4', 'Värillinen_A4', 'Mustavalkoinen_A3', 'Värillinen_A3', 'Yhteensä']
    variables = ['Mustavalkoinen_A4', 'Värillinen_A4', 'Mustavalkoinen_A3', 'Värillinen_A3', 'Yhteensä']
    unit_means_data = printing_data[unit_means_columns]

    unit_means = unit_means_data.groupby('Toimintayksikkö')[variables].mean()

    min_values = unit_means.min()
    min_units = unit_means.idxmin()
    max_values = unit_means.max()
    max_units = unit_means.idxmax()

    min_freq = unit_means.apply(lambda x: x.value_counts().get(min_values[x.name], 0))
    max_freq = unit_means.apply(lambda x: x.value_counts().get(max_values[x.name], 0))

    unit_means_extremes = pd.DataFrame({
        'Mean_min': min_values,
        'Toimintayksikkö_min': min_units,
        'Mean_max': max_values,
        'Toimintayksikkö_max': max_units,
        'Erotus': max_values - min_values
    }, index=variables)

    unit_means_extremes['Toimintayksikkö_min'] = unit_means_extremes.apply(lambda row: f"{min_freq[row.name]} toimintayksikköä" if min_freq[row.name] > 1 else row['Toimintayksikkö_min'], axis=1)
    unit_means_extremes['Toimintayksikkö_max'] = unit_means_extremes.apply(lambda row: f"{max_freq[row.name]} toimintayksikköä" if max_freq[row.name] > 1 else row['Toimintayksikkö_max'], axis=1)

    display(unit_means_extremes)

    

    # Creating a logarithm based visualization for comparing min and max MEAN values of print types and total sum
    
    filtered_min_values = min_values.replace(0, np.nan)
    filtered_max_values = max_values.replace(0, np.nan)

    log_min_values = np.log10(filtered_min_values)
    log_max_values = np.log10(filtered_max_values)

    fig, ax = plt.subplots()

    ax.plot(variables, log_min_values, marker='o', label='Minimi', color='lightblue')
    ax.plot(variables, log_max_values, marker='o', label='Maksimi', color='darkblue')

    ax.set_title('Toimintayksiköiden keskiarvojen logaritmiset minimi- ja maksimiarvot')
    ax.set_xlabel('Muuttuja')
    ax.set_ylabel('Logaritmi')
    ax.legend()

    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.show()

```
